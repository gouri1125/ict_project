```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**IRIS-DATA VISHUALISATION**

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris_new=iris.select_dtypes(include=['number'])
corr=iris_new.corr()
corr
```

**HEATMAP**

```python
sns.heatmap(corr, annot=True, cmap='Greens')
```

```python
sns.pairplot(iris,hue='species')
```

```python
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.boxplot(data=iris,x="petal_length",hue="species")
```

```python
iris_species=iris['species'].value_counts()
labels=iris_species.index
plt.pie(iris_species,labels=labels)
plt.show()
```

```python
sns.histplot(data=iris,x="petal_width",hue="species",kde=True)
```

```python
sns.kdeplot(iris)
```

```python
sns.jointplot(data=iris,x="petal_length",y="petal_width",hue="species")#combines scatter + histogram (Two Variables + Distribution)
```

```python
sns.countplot( data=iris, x="species")#show number of samples
```

```python
sns.violinplot( data=iris, x="species", y="petal_length")
```

```python
sns.ecdfplot(
    data=iris,
    x="petal_length",
    hue="species")
```
