8) Handle missing values in 'Revenue (Millions)' and 'Metascore

```python
revenue_mean = df['Revenue (Millions)'].mean()
metascore_mean = df['Metascore'].mean()

df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(revenue_mean)
df['Metascore'] = df['Metascore'].fillna(metascore_mean)

print("Revenue Mean:", revenue_mean)
print("Metascore Mean:", metascore_mean,'\n')
print(df['Revenue (Millions)'],['Metascore'])
```

17) Create a column that categorizes movies by rating:
a) Rating < 6: Poor
b) 6 â‰¤ Rating < 7.5: Average
c) Rating â‰¥ 7.5: Excellent

```python
df['Rating_Category'] = np.where(df['Rating'] < 6, 'Poor',
                        np.where(df['Rating'] < 7.5, 'Average', 'Excellent'))
df['Rating_Category']
```

18) Create a custom function to score movies based on:
i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
ii)Apply a function to dataset to assign the score

```python
def get_high_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']
df['High_Score'] = df.apply(get_high_score, axis=1)
df['High_Score']
```
