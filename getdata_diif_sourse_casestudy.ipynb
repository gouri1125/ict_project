```python
import pandas as pd
import numpy as np
```

Step 1: Load SpaceX Launch Data from API

```python
url1 =pd.read_json('https://api.spacexdata.com/v4/launches')
url1
```

```python
url1.info()
```

```python
#Extract relevant columns: name, date_utc, success, details, rocket
relevent_column = url1[['name','date_utc','success','details','rocket']]
relevent_column.head()
```

```python
relevent_column = relevent_column.copy()
# Convert date_utc to datetime
relevent_column['datetime'] = pd.to_datetime(relevent_column['date_utc'], errors='coerce')

# Extract year
relevent_column['year'] = relevent_column['datetime'].dt.year
relevent_column
```

Step 2: Load Rocket Metadata

```python
url2 =pd.read_json('https://api.spacexdata.com/v4/rockets')
url2
```

```python
#Extract id, name, type, active, and stages
selected_column = url2[['id','name','type','active','stages']]
selected_column.head()
```

Step 3: Merge Launch and Rocket Data


```python
#Join the two DataFrames on rocket ID using
mergr2 = selected_column.set_index('id').add_suffix('_rocket')
merged_table = relevent_column.join(mergr2, on='rocket')
# Show result
merged_table.head()
```

Step 4: Add Simulated Country Information


```python
# Add a new column country and randomly assign one of these values:
#['USA', 'Russia', 'India', 'China', 'France']

countries = ['USA', 'Russia', 'India', 'China', 'France']
# Add new column with random assignment
merged_table['country'] = np.random.choice(countries, size=len(merged_table))
merged_table
```

Step 5: Store Merged Data in SQLite3

```python
#Use sqlite3 to create a connection and save the merged DataFrame as a table named launches
#Table should contain all merged columns including country

import sqlite3
# Create SQLite connection (creates file if not exists)
conn = sqlite3.connect("spacex.db")

# Save merged db
merged_table.to_sql("launches", conn, if_exists="replace", index=False)
merged_table
```

Step 6: Run SQL Queries on the Data to analyze

```python
#1. Launches by Country
query = """
SELECT country, COUNT(*) AS launch_count
FROM launches
GROUP BY country
ORDER BY launch_count DESC
"""
pd.read_sql_query(query, conn)
```

```python
#2. Which year had the highest number of launches?
query = """
SELECT year, COUNT(*) AS launch_count
FROM launches
GROUP BY year
ORDER BY launch_count DESC LIMIT 1
"""
pd.read_sql_query(query, conn)
```

```python
#3. Top 5 Missions by Launch Count
query = """
SELECT name AS mission_name,
       COUNT(*) AS launch_count
FROM launches
GROUP BY mission_name
ORDER BY launch_count DESC
LIMIT 5;
"""
pd.read_sql_query(query, conn)
```

```python
#4. top 5 rockets by number of launch
query = """
SELECT name_rocket AS rocket_name,
       COUNT(*) AS launch_count
FROM launches
GROUP BY rocket_name
ORDER BY launch_count DESC
LIMIT 5;
"""

pd.read_sql_query(query, conn)
```
