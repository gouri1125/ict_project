outlier removal

```python
num_cols = houseprice.select_dtypes(include=['int64', 'float64']).columns

outlier_summary = {}

for col in num_cols :
    Q1 = houseprice[col].quantile(0.25)
    Q3 = houseprice[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = houseprice[(houseprice[col] < lower) | (houseprice[col] > upper)]
    outlier_summary[col] = outliers.shape[0]

outlier_summary
```

```python
cols = [
    'No of Bedrooms',
    'No of Bathrooms',
    'Flat Area (in Sqft)',
    'Lot Area (in Sqft)',
    'No of Floors',
    'Area of the House from Basement (in Sqft)',
    'Basement Area (in Sqft)',
    'Age of House (in Years)',
    'Renovated Year',
    'Living Area after Renovation (in Sqft)',
    'Lot Area after Renovation (in Sqft)'
]
for col in cols:
    Q1 = houseprice[col].quantile(0.25)
    Q3 =houseprice[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    houseprice[col] = houseprice[col].clip(lower_limit, upper_limit)
houseprice
```
